```python
import pandas as pd
import numpy as np
```

```python
Cdata= pd.read_excel("7.-Medicine-Data-with-lumped-Quantity-and-Measure.xlsx")
```

```python
Cdata
```

```python
# '(\d)' represents digit while {1,6} 
extracted_figures = Cdata['Quantity'].str.extract('(\d{1,6})', expand=True)
```

```python
extracted_figures
```

```python
#Extracting the text
extracted_text=  Cdata['Quantity'].str.strip('\d{1,2000}')
```

```python
extracted_text= pd.DataFrame(extracted_text)
extracted_text
```

```python
#renaming the column of extracted_text
extracted_text.rename(columns = {'Quantity':'Measure'})
```

```python
#joining the measure column to Cdata
Cdata['Measure'] = extracted_text
```

```python
#joining the Quality column to Cdata
Cdata['Quantity']= extracted_figures
Cdata
```
