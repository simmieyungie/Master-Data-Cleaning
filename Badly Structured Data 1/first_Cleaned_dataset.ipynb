```python
import pandas as pd
import numpy as np
```

```python
#Read in the Data
Cdata = pd.read_csv("main.csv")
```

```python
#Checking the Data structure
Cdata.head()
```

```python
#Renaming the Columns
Cdata.rename(columns= {'Segment>>':'Order ID'}, inplace = True)
Cdata.rename(columns = {'Unnamed: 12':'Home Office'}, inplace = True)
Cdata.rename(columns = {'Unnamed: 13':'Home Office'}, inplace = True)
Cdata.rename(columns = {'Unnamed: 14':'Home Office'}, inplace = True)
Cdata.rename(columns = {'Unnamed: 2':'Consumer', 'Unnamed: 3':'Consumer', 'Unnamed: 4':'Consumer'}, inplace = True)
Cdata.rename(columns = {'Unnamed: 7':'Corporate', 'Unnamed: 8':'Corporate', 'Unnamed: 9':'Corporate'}, inplace = True)
```

```python
#Extracting the data to edit the first part (Clean Data 1 = Cdata1)
Cdata1 = Cdata.iloc[0:-1]
Cdata1
```

```python
#Editing the Rows
Cdata1.drop(0, inplace = True)
Cdata1.drop(1, inplace = True)
Cdata1
```

```python
#Droping unwanted Columns
Cdata1.drop(['Consumer Total','Corporate Total', 'Home Office Total' ], axis= 1 , inplace =True)
Cdata1
```

```python
#Resetting the Index 
Cdata1.reset_index(drop=True,inplace=True)
Cdata1
```

```python
#Turning the Coulunms into rows 
Cdata1= pd.melt(Cdata1, id_vars=['Order ID'], value_vars=['Consumer','Consumer','Consumer', 'Consumer','Corporate',
                                                           'Corporate','Corporate', 'Corporate','Corporate',
                                                           'Home Office','Home Office', 'Home Office','Home Office'],
                                                            var_name='Segment', value_name='Sales')
```

```python
#We have the First Phase Cleaned
Cdata1
```

```python
#Extracting the data to edit the Second Phase (Clean Data 2 = Cdata2)
Cdata2 =Cdata.iloc[0:-1]
```

```python
#Drop unwanted Columns
Cdata2.drop(['Consumer Total','Corporate Total', 'Home Office Total' ], axis= 1 , inplace =True)
```

```python
#Making the First Row the New Column
Cdata2.columns= Cdata2.iloc[0]
Cdata2
```

```python
#Droping the unwanted rows
Cdata2.drop(0, inplace = True)
Cdata2.drop(1, inplace = True)
Cdata2
```

```python
#Resetting the Index
Cdata2.reset_index(drop=True,inplace=True)
```

```python
Cdata2.columns
```

```python
#Renaming the Column
Cdata2.rename(columns= {'Ship Mode>>':'Order ID'}, inplace = True)
```

```python
Cdata2.columns
```

```python
#Turning the columns into rows
Cdata2 = pd.melt(Cdata2, id_vars=['Order ID'], value_vars= ['First Class', 'Same Day', 'Second Class', 'Standard Class',
       'First Class', 'Same Day', 'Second Class', 'Standard Class',
       'First Class', 'Same Day', 'Second Class', 'Standard Class'] ,
                        var_name='Ship Mode',    value_name='Sales')
Cdata2
```

```python
#Merging the two phases of Data together 
Full = pd.merge(Cdata1, Cdata2, on = ('Order ID', 'Sales'))
```

```python
#Droping Missing Values
Full.dropna(inplace = True )
```

```python
#Droping Duplicate Values and leaving the
Full.drop_duplicates(subset ="Order ID", 
                     keep = 'first', inplace = True) 
```

```python
#Reseting the index
Full.reset_index(drop=True,inplace=True)
```

```python
#Data Cleaned
First_cleaned=Full
```
