```python
import pandas as pd
import numpy as np
```

```python
Cdata = pd.read_excel('2.-Badly-Structured-Sales-Data-2.xlsx')
```

```python
Cdata
```

```python
Cdata.columns
```

```python
#Renaming the columns
Cdata.rename(columns = {'Ship Mode': 'Order Date' }, inplace = True)
Cdata.rename(columns = {'Unnamed: 2': 'First Class' }, inplace = True)
Cdata.rename(columns = {'Unnamed: 3': 'First Class' }, inplace = True)
Cdata.rename(columns = {'Unnamed: 5': 'Same Day' }, inplace = True)
Cdata.rename(columns = {'Unnamed: 6': 'Same Day' }, inplace = True)
Cdata.rename(columns = {'Unnamed: 8': 'Second Class' }, inplace = True)
Cdata.rename(columns = {'Unnamed: 9': 'Second Class' }, inplace = True)
Cdata.rename(columns = {'Unnamed: 11': 'Standard Class' }, inplace = True)
Cdata.rename(columns = {'Unnamed: 12': 'Standard Class' }, inplace = True)
```

```python
Cdata
```

```python
#extracting the first phase of the data to be cleaned
Cdata1 = Cdata.iloc[2:]
```

```python
Cdata1
```

```python
#resetting the index
Cdata1.reset_index(drop= True, inplace = True)
```

```python
#turning the columns into row
Cdata1 = pd.melt(Cdata1 , id_vars= ['Order Date'] , value_vars= ['First Class', 'First Class', 'First Class', 'Same Day',
       'Same Day', 'Same Day', 'Second Class', 'Second Class', 'Second Class',
       'Standard Class', 'Standard Class', 'Standard Class'], var_name = "Ship Mode", value_name = "Sales" )
```

```python
Cdata1
```

```python
#extracting the second phase 
Cdata2 =Cdata.iloc[0:]
```

```python
#making the first row the new colunm
Cdata2.columns= Cdata2.iloc[0]
```

```python
#dropping the wanted rows
Cdata2.drop(0, inplace = True)
Cdata2.drop(1, inplace = True)
```

```python
Cdata2.columns
```

```python
#Renaming the column
Cdata2.rename(columns= {'Segment':'Order Date'}, inplace = True)
```

```python
#Resetting the index
Cdata2.reset_index(drop=True,inplace=True)
```

```python
Cdata2 =  pd.melt(Cdata2 , id_vars= ['Order Date'] , value_vars= ['Consumer', 'Corporate', 'Home Office', 'Consumer',
       'Corporate', 'Home Office', 'Consumer', 'Corporate', 'Home Office',
       'Consumer', 'Corporate', 'Home Office'], var_name = "Segment", value_name = "Sales" )
```

```python
Cdata2
```

```python
#merging the two cleaned phases
Full = pd.merge(Cdata1, Cdata2, on = ('Order Date', 'Sales'))
```

```python
Full
```

```python
#Dropping null values
Full.dropna(inplace = True )
```

```python
#Droppping the duplicate but keeping thefirst
Full.drop_duplicates(subset =["Sales", "Order Date"], 
                     keep = 'first', inplace = True) 
```

```python
#resetting the index
Full.reset_index(drop=True,inplace=True)
```

```python
#saving file as csv
Full.to_csv('Second_Cleaned.csv', index=False)
```
